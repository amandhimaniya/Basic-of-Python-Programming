```python
# Write a program to find the factorial of any number using the concept of recursion.
def fact(n):
    if n==1:
        return(1)
    else:
        return (n*fact(n-1))
fact(6)
```

```python
# Write a Python function that takes a number as a parameter and check if the number is prime or not.
def Prime():
    n=int(input("Enter the number which you checked is it prime or not :- "))
    if n==1 :
        print("1 is not prime")
    else :
        for i in range(2,11):
            if n%i==0:
                print("The number is not Prime")
                break
            else:
                print("The number is  prime")
                break
            
    
Prime()
```

```python
# Write a Python function to sum all the numbers in a list using Recursion
def Sum(number):
    if len(number) == 1:
        return number[0]
    else:
        return number[0]+Sum(number[1:])
        
Sum([10,11,22,45,84,95,33])
```
